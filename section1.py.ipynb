```python
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]
```

```python
#a)
ages.sort()
print("Sorted ages:", ages)

min_age = min(ages)
max_age = max(ages)

print("Minimum age:", min_age)
print("Maximum age:", max_age)
```

```python
#b)
ages.append(min_age)
ages.append(max_age)
print("Ages after adding min and max again:", ages)
```

```python
#c)
ages.sort()  # Sort again after appending
length = len(ages)

if length % 2 == 0:
    # Even number of elements — average the two middle values
    median = (ages[length//2 - 1] + ages[length//2]) / 2
else:
    # Odd number of elements — take the middle value
    median = ages[length//2]

print("Median age:", median)
```

```python
#a)
score = int(input("Enter the student's score (0-100): "))

if 80 <= score <= 100:
    grade = 'A'
elif 70 <= score <= 79:
    grade = 'B'
elif 60 <= score <= 69:
    grade = 'C'
elif 50 <= score <= 59:
    grade = 'D'
elif 0 <= score <= 49:
    grade = 'F'
else:
    grade = 'Invalid score'

print("Grade:", grade)
```

```python
#c)
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):  # Only check up to sqrt(n)
        if n % i == 0:
            return False
    return True


print("Is 7 prime?", is_prime(7))    # True
print("Is 10 prime?", is_prime(10))  # False
```
